# Обзор библиотек Matplotlib, Seaborn и Plotly

### **Matplotlib**— одна из наиболее популярных библиотек для визуализации данных на Python. Она предлагает широкий набор инструментов для построения различных графиков и диаграмм.

## Преимущества Matplotlib:

•	Гибкость. Библиотека позволяет настраивать почти все аспекты графика, чтобы создать точно тот вид визуализации, который необходим.

•	Разнообразие типов графиков. Matplotlib предлагает широкий спектр типов графиков, таких как линейные графики, гистограммы, диаграммы разброса, контурные графики, трёхмерные визуализации и многое другое.

•	Простота использования. Благодаря простому синтаксису создание графиков с помощью Matplotlib становится доступным даже для новичков.  

•	Кастомизация графиков. Библиотека предоставляет широкие возможности по настройке внешнего вида графиков, включая цвета, шрифты, размеры элементов и другие параметры.

•	Интерактивные возможности. Matplotlib поддерживает интерактивные функции, такие как масштабирование, перемещение, выделение и сохранение графиков в различных форматах.

•	Совместимость. Matplotlib работает везде, где есть Python.

•	Обширная документация.  Matplotlib располагает исчерпывающей документацией и большим сообществом пользователей и разработчиков, что упрощает поиск справки, руководств и примеров.

## Недостатки Matplotlib:

•	Крутая кривая обучения. Для начинающих пользователей это может быть сложно из-за обширных возможностей настройки и иногда сложного синтаксиса.

•	Подробный синтаксис. Он может быть менее интуитивно понятным по сравнению с другими библиотеками построения графиков, что отнимает больше времени на создание и настройку графиков.  

•	Эстетика по умолчанию. Графики по умолчанию в Matplotlib часто считаются менее визуально привлекательными по сравнению с другими библиотеками, что требует больше усилий для придания графикам привлекательности.

•	Ограниченная интерактивность. Хотя Matplotlib в некоторой степени поддерживает интерактивное построение графиков, он не предлагает такого количества интерактивных функций и опций, как другие библиотеки.

•	Ограниченные возможности 3D-графики. Возможности 3D-графики Matplotlib не такие продвинутые и удобные для пользователя, как у некоторых других специализированных библиотек 3D-графики.  

•	Проблемы с производительностью при работе с большими наборами данных. Matplotlib иногда может быть медленнее и менее эффективен при построении графиков больших наборов данных, особенно по сравнению с более оптимизированными библиотеками построения графиков.

•	Зависимость от внешних библиотек. Matplotlib полагается на другие библиотеки, такие как NumPy и SciPy, для многих своих функциональных возможностей, что иногда может приводить к проблемам совместимости и управления зависимостями.

•	Ограниченная встроенная поддержка построения статистических графиков. Хотя Matplotlib может создавать базовые статистические графики, в нём отсутствуют некоторые расширенные возможности построения статистических графиков, которые доступны в специализированных библиотеках.

•	Менее современные функции. Matplotlib существует уже давно, и некоторые пользователи обнаруживают, что в нём отсутствуют некоторые современные функции построения графиков и интерактивной визуализации, имеющиеся в более новых библиотеках.  


### **Seaborn** — это библиотека, основанная на Matplotlib, которая предоставляет дополнительные функции и улучшенный дизайн графиков. Seaborn упрощает создание сложных визуализаций и предлагает встроенные темы, которые делают графики более привлекательными.  

## Преимущества Seaborn:

•  Богатая коллекция готовых к использованию типов графиков. Это позволяет визуализировать распределения, корреляции и тренды, не погружаясь в математические детали.

•  Встроенная поддержка для создания многослойных графиков с использованием категориальных переменных. Библиотека автоматически справляется с группировкой данных и настройкой циклов для создания сложных графиков.

•  Контроль над эстетикой графика. В Seaborn есть множество встроенных тем и стилей, которые позволяют улучшить внешний вид графиков. Это включает масштабирование элементов и кастомизацию цветовых палитр.

•  Улучшенные возможности для многомерного анализа. С помощью функций, таких как pairplot и heatmap, можно исследовать парные отношения и корреляции между несколькими переменными.

•  Простота использования и высокая гибкость. Библиотека предоставляет простой и понятный синтаксис для создания графиков. Несмотря на свою простоту, позволяет детально настраивать внешний вид графиков.

•  Интеграция с pandas. Библиотека позволяет напрямую передавать табличные данные pandas DataFrame в функции построения графиков, что значительно ускоряет процесс визуализации.


## Недостатки Seaborn:

•  Зависимость от Matplotlib. Поскольку Seaborn построена на базе Matplotlib, её возможности ограничены функциональностью этой библиотеки.

•  Кривая обучения для продвинутой настройки. Хотя Seaborn легка в освоении для создания стандартных визуализаций, более сложная кастомизация графиков может потребовать понимания как этой библиотеки, так и Matplotlib.

•  Ограниченная поддержка 3D-графиков и интерактивности. Seaborn сосредоточена на создании статических 2D-графиков и не предназначена для работы с трёхмерной визуализацией или интерактивными графиками.

•  Производительность с крупными объёмами данных. При работе с очень большими объёмами данных Seaborn может оказаться менее производительной по сравнению с некоторыми другими инструментами.

•  Дефицит документации для сложных случаев. Хотя базовая документация Seaborn является чёткой и хорошо организованной, примеры и руководства для более сложных сценариев использования могут быть недостаточно подробными.

•  Меньшее сообщество по сравнению с другими инструментами. Seaborn имеет относительно маленькое сообщество по сравнению с более широко используемыми инструментами, такими как Matplotlib или Plotly.


### **Plotly** — это библиотека для создания интерактивных визуализаций данных. Она позволяет создавать графики, которые пользователи могут исследовать, изменять и делиться с другими. Plotly поддерживает множество типов графиков, включая линейные графики, столбчатые диаграммы, круговые диаграммы и тепловые карты.  

## Преимущества Plotly:

•  Интерактивность. Графики и диаграммы реагируют на действия пользователя, такие как наведение курсора, щелчок и масштабирование.

•  Разнообразие графиков. Библиотека содержит множество типов графиков, включая линейные графики, гистограммы, точечные диаграммы, контурные карты, трёхмерные графики и т. д..

•  Богатая настройка графиков. Пользователи могут управлять различными аспектами — цветовые палитры, шрифты, оси, аннотации, легенды, заголовки и макет.

•  Поддержка нескольких платформ. Plotly поддерживает различные платформы и языки программирования, включая Python, JavaScript, R и MATLAB.

•  Экспорт и встраивание. Графики, созданные с помощью Plotly, можно экспортировать в различные форматы (например, PNG, SVG) и встраивать в веб-страницы, блокноты Jupyter и другие приложения.

•  Совместная работа и обмен данными в интернете. Plotly предлагает онлайн-платформу Plotly Chart Studio, где пользователи могут создавать, редактировать и делиться графиками.


## Недостатки Plotly:

•  Низкая производительность при работе с большими датасетами. Рендеринг таких данных может быть медленным по сравнению с Matplotlib и Seaborn.

•  Сложности с освоением. Несмотря на простоту использования для базовых графиков, освоить все возможности Plotly, особенно низкоуровневый API объектов графиков, может быть сложно.

•  Зависимость от веб-технологий. Интерактивные графики в Plotly отображаются с использованием веб-технологий, таких как HTML, CSS и JavaScript. Это может быть недостатком в средах, где веб-доступ ограничен или недоступен.

•  Размер и сложность файлов. Интерактивные графики могут приводить к увеличению размера файлов и более сложным HTML-документам. Это может быть проблемой при обмене результатами или встраивании графиков на веб-страницы, где важны производительность и простота.

•  Ограниченные возможности настройки по сравнению с Matplotlib. В Matplotlib есть широкие возможности настройки, которые не всегда доступны в Plotly.

# Сравнение библиотек Matplotlib, Seaborn и Plotly на примере csv файла 'Преступность в России: 2003 – 2020 годы'

## Визуализация с использованием Matplotlib:

```python
import pandas as pd
import matplotlib.pyplot as plt
import mplcyberpunk


data = pd.read_csv('crime_data.csv')  # путь файлу

# столбец 'month' в формате datetime
data['month'] = pd.to_datetime(data['month'])

# Установка стиля mplcyberpunk
plt.style.use('cyberpunk')

# Визуализация общего количества преступлений по месяцам
plt.figure(figsize=(14, 7))
plt.plot(data['month'], data['Total_crimes'], label='Total Crimes', color='cyan', linewidth=2)
plt.title('Общее количество преступлений в России (2003-2020)', fontsize=16)
plt.xlabel('Месяц', fontsize=14)
plt.ylabel('Количество преступлений', fontsize=14)
plt.legend()
plt.grid()
mplcyberpunk.add_glow_effects()
plt.show()
```

```python
# Визуализация серьезных преступлений
plt.figure(figsize=(14, 7))
plt.plot(data['month'], data['Serious'], label='Серьезные преступления', color='magenta', linewidth=2)
plt.title('Серьезные преступления в России (2003-2020)', fontsize=16)
plt.xlabel('Месяц', fontsize=14)
plt.ylabel('Количество серьезных преступлений', fontsize=14)
plt.legend()
plt.grid()
mplcyberpunk.add_glow_effects()
plt.show()
```

```python
# Визуализация убийств
plt.figure(figsize=(14, 7))
plt.plot(data['month'], data['Murder'], label='Убийства', color='red', linewidth=2)
plt.title('Количество убийств в России (2003-2020)', fontsize=16)
plt.xlabel('Месяц', fontsize=14)
plt.ylabel('Количество убийств', fontsize=14)
plt.legend()
plt.grid()
mplcyberpunk.add_glow_effects()
plt.show()
```

## Визуализация с использованием Seaborn:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('crime_data.csv')  # путь к файлу

# Преобразование столбца 'month' в формат даты
data['month'] = pd.to_datetime(data['month'])

# Установка стиля Seaborn
sns.set(style="whitegrid")

# Визуализация общего количества преступлений по месяцам
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x='month', y='Total_crimes', marker='o')
plt.title('Общее количество преступлений в России (2003-2020)')
plt.xlabel('Месяц')
plt.ylabel('Количество преступлений')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Визуализация серьезных преступлений
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x='month', y='Serious', marker='o', color='red')
plt.title('Серьезные преступления в России (2003-2020)')
plt.xlabel('Месяц')
plt.ylabel('Количество серьезных преступлений')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Визуализация различных типов преступлений
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x='month', y='Murder', marker='o', label='Убийства')
sns.lineplot(data=data, x='month', y='Rape', marker='o', label='Изнасилования')
sns.lineplot(data=data, x='month', y='Theft', marker='o', label='Кражи')
plt.title('Различные типы преступлений в России (2003-2020)')
plt.xlabel('Месяц')
plt.ylabel('Количество преступлений')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Скрипичный график
data['month'] = pd.to_datetime(data['month'])
data['year'] = data['month'].dt.year
plt.figure(figsize=(12, 6))
sns.violinplot(x='year', y='Total_crimes', data=data)
plt.title('Скрипичный график состояния преступности по годам')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Визуализация с использованием Plotly:

```python
import pandas as pd
import plotly.express as px
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv('crime_data.csv')

# Преобразование колонки 'month' в формат datetime
data['month'] = pd.to_datetime(data['month'])

# Линейный график общего количества преступлений по месяцам
fig1 = px.line(data, x='month', y='Total_crimes', title='Общее количество преступлений в России (2003-2020)',
               labels={'Total_crimes': 'Количество преступлений'},
               template='plotly_white')

fig1.show()
```

```python
# Суммирование данных по типам преступлений
crime_types = data[['Serious', 'Huge_damage', 'Ecological', 'Terrorism',
                    'Extremism',
                    'Murder', 'Harm_to_health', 'Rape', 'Theft', 'Vehicle_theft',
                    'Fraud_scam', 'Hooligan', 'Drugs', 'Weapons']].sum()

# Создание круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(crime_types, labels=crime_types.index, autopct='%1.1f%%', startangle=140)
plt.title('Распределение типов преступлений в России (2003-2020)')
plt.axis('equal')  # Чтобы круг был кругом
plt.show()
```

# Рекомендации по выбору между Matplotlib, Seaborn и Plotly:

**Matplotlib** подходит для высококачественных статических графиков и сложных
визуализаций. Библиотека гибкая в кастомизации графиков, поддерживает различные типы визуализаций.

**Seaborn** рекомендуется для удобного создания статистических графиков и работы с Pandas. Библиотека построена на основе Matplotlib, имеет встроенные стили и палитры.

**Plotly** подходит для интерактивных графиков и визуализаций, требующих взаимодействия с пользователем. Библиотека поддерживает интерактивность и анимацию, имеет возможность создания 3D-визуализаций.

Таким образом, Matplotlib подходит для статических графиков, Seaborn — для статистических визуализаций, а Plotly — для интерактивных графиков.

Не стоит забывать что выбор используемой библиотеки зависит от конкретных задач и предпочтений разработчика.





# Заключение

Matplotlib является мощным инструментом для создания визуализаций данных и подходит для широкого спектра задач, от простых графиков до сложных научных визуализаций.

Seaborn является отличным инструментом для статистической визуализации и анализа категориальных данных. Эта библиотека подходит для большинства задач визуализации, особенно когда важны эстетика и понятность графиков.

Plotly является отличным выбором для создания интерактивных и сложных визуализаций, особенно когда важна возможность взаимодействия с данными. Эта библиотека подходит для различных задач, включая анализ больших объемов данных, веб-визуализацию и работу с геоданными.
